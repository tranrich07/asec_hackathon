```python
ds = df.drop(['time'], axis=1)

X = ds.values
train_size = int(len(X) * 0.5)
```

```python
# Granger's Causality Test
maxlag=12
test = 'ssr_chi2test'

def grangers_causation_matrix(data, variables, test='ssr_chi2test', verbose=False):    
    df = pd.DataFrame(np.zeros((len(variables), len(variables))), columns=variables, index=variables)
    for c in df.columns:
        for r in df.index:
            test_result = grangercausalitytests(data[[r, c]], maxlag=maxlag, verbose=False)
            p_values = [round(test_result[i+1][0][test][1],4) for i in range(maxlag)]
            if verbose: print(f'Y = {r}, X = {c}, P Values = {p_values}')
            min_p_value = np.min(p_values)
            df.loc[r, c] = min_p_value
    df.columns = [var + '_x' for var in variables]
    df.index = [var + '_y' for var in variables]
    return df

grangers_causation_matrix(ds, variables = ds.columns)  
```

```python
# Data Partitioning
n_obs = 4
ds_train, ds_test = ds[0:-n_obs], ds[-n_obs:]

# Check size
print(ds_train.shape)  # (119, 8)
print(ds_test.shape)  # (4, 8)
```

```python
# Augmented Dickey-Fuller Test

def adf_test(timeseries):
    print('Results: ')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags','# Observations'])
    
    for key,value in dftest[4].items():
       dfoutput['Critical Value (%s)'%key] = value
    print (dfoutput)
    
    p = dfoutput['p-value']
    if p <= 0.05:
        print("\nSeries is Stationary")
    else:
        print("\nSeries is Non-Stationary")
        
# Testing for statonary
for col in ds_train.columns:
    print('Columns: ', col)
    print('------------------------------------------')
    adf_test(ds_train[col])
    print('\n')
    
# Get rid of stationary variables
ds_differentiated = ds_train.diff().dropna()

# Run test again
for col in ds_train.columns:
    print('Columns: ', col)
    print('------------------------------------------')
    adf_test(ds_differentiated[col])
    print('\n')
```

```python
# Vector Autoregression (VAR) model

model = VAR(ds_differentiated)
results = model.fit(maxlags = 15, ic = 'aic')
# results.summary()
```

```python
# Forecasting
lag_order = results.k_ar
predicted = results.forecast(ds_differentiated.values[-lag_order:], n_obs)
forecast = pd.DataFrame(predicted, index = ds.index[-n_obs:], columns = ds.columns)
```

```python
# Inverting the Differencing Transformation
def invert_transformation(ds, df_forecast, second_diff=False):
    for col in ds.columns:
        # Undo the 2nd Differencing
        if second_diff:
            df_forecast[str(col)] = (ds[col].iloc[-1] - ds[col].iloc[-2]) + df_forecast[str(col)].cumsum()

        # Undo the 1st Differencing
        df_forecast[str(col)] = ds[col].iloc[-1] + df_forecast[str(col)].cumsum()

    return df_forecast

forecast_values = invert_transformation(ds_train, forecast, second_diff=False)
forecast_values
```

```python
# MSE
mse = mean_squared_error(ds_test, forecast_values)
print(mse)
```
